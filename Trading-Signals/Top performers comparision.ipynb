```python
import os
import pandas as pd

# Function to read all Excel files from a folder and add a 'Strategy' column to each DataFrame
def read_files_from_folder(folder_path):
    all_data = []  # List to store all the dataframes

    # Loop through each file in the folder
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        
        # Check if the file is an Excel file
        if file_name.endswith('.xlsx'):
            try:
                # Try to read the 'Top' and 'Bottom' sheets from the file
                excel_file = pd.ExcelFile(file_path)
                
                # Check if both 'Top' and 'Bottom' sheets exist
                if 'Top 10' in excel_file.sheet_names and 'Bottom 10' in excel_file.sheet_names:
                    top_data = excel_file.parse('Top 10')
                    bottom_data = excel_file.parse('Bottom 10')
                    
                    # Use full filename (without extension) as the strategy name
                    strategy_name = file_name.split('.')[0]  # Use the whole filename without the extension
                    top_data['Strategy'] = strategy_name
                    bottom_data['Strategy'] = strategy_name
                    
                    # Append the data from 'Top' and 'Bottom' sheets to the list
                    all_data.append(top_data)
                    all_data.append(bottom_data)
                else:
                    print(f"Skipping {file_name}: Missing 'Top' or 'Bottom' sheet.")
            except Exception as e:
                print(f"Error reading {file_name}: {e}")

    # Return a combined dataframe containing data from all valid files
    if all_data:
        return pd.concat(all_data, ignore_index=True)
    else:
        return pd.DataFrame()  # Return an empty DataFrame if no valid data was found


# Function to rank strategies by Sharpe Ratio and select the top and bottom 30%
def rank_strategies(data):
    # Sort by Sharpe Ratio (descending)
    sorted_data = data.sort_values(by='Sharpe Ratio', ascending=False)
    top_30_percent = sorted_data.head(int(len(sorted_data) * 0.3))  # Top 30% of data
    bottom_30_percent = sorted_data.tail(int(len(sorted_data) * 0.3))  # Bottom 30% of data

    return top_30_percent, bottom_30_percent


# Function to find consistent top and bottom performers across train and test datasets
def find_consistent_performers(train_data, test_data):
    # Rank both train and test data
    top_train, bottom_train = rank_strategies(train_data)
    top_test, bottom_test = rank_strategies(test_data)
    
    # Count occurrences of each strategy in the top 30% of both train and test
    top_train_counts = top_train['Strategy'].value_counts()
    top_test_counts = top_test['Strategy'].value_counts()
    
    # Count occurrences of each strategy in the bottom 30% of both train and test
    bottom_train_counts = bottom_train['Strategy'].value_counts()
    bottom_test_counts = bottom_test['Strategy'].value_counts()

    # Merge counts from train and test
    performance_data = pd.DataFrame({
        'Top Train': top_train_counts,
        'Top Test': top_test_counts,
        'Bottom Train': bottom_train_counts,
        'Bottom Test': bottom_test_counts
    }).fillna(0)

    # Add a column that calculates the "Total Occurrence" in top 30% and bottom 30%
    performance_data['Top Consistency'] = performance_data['Top Train'] + performance_data['Top Test']
    performance_data['Bottom Consistency'] = performance_data['Bottom Train'] + performance_data['Bottom Test']

    # Rank the strategies by "Top Consistency" (descending) and "Bottom Consistency" (ascending)
    top_performers = performance_data.sort_values(by='Top Consistency', ascending=False)
    bottom_performers = performance_data.sort_values(by='Bottom Consistency', ascending=False)

    return top_performers, bottom_performers


# Function to process all files, identify consistent performers, and save results
def process_results(train_folder, test_folder, output_file):
    # Read files from both train and test folders
    print("Reading train data...")
    train_data = read_files_from_folder(train_folder)
    print("Reading test data...")
    test_data = read_files_from_folder(test_folder)
    
    # Find consistent top and bottom performers
    print("Finding consistent performers...")
    top_performers, bottom_performers = find_consistent_performers(train_data, test_data)
    
    # Save results to an Excel file
    print(f"Saving results to {output_file}...")
    with pd.ExcelWriter(output_file) as writer:
        top_performers.to_excel(writer, sheet_name='Consistent Top Performers', index=True)
        bottom_performers.to_excel(writer, sheet_name='Consistent Bottom Performers', index=True)
    
    print("Process completed and results saved.")


# Main execution
if __name__ == "__main__":
    # Define the train and test folder paths
    train_folder = r"C:\Users\masho\OneDrive\Documents\Fall 24 SEM\Capstone\Final Results\BTC 2024 15min TR"
    test_folder = r"C:\Users\masho\OneDrive\Documents\Fall 24 SEM\Capstone\Final Results\BTC 2024 15min T"
    
    # Output file name
    output_file = 'consistent_performers.xlsx'
    
    # Run the process
    process_results(train_folder, test_folder, output_file)
```

```python
import os
import pandas as pd

# Function to read all Excel files from a folder and add a 'Strategy' column to each DataFrame
def read_files_from_folder(folder_path):
    all_data = []  # List to store all the dataframes

    # Loop through each file in the folder
    for file_name in os.listdir(folder_path):
        file_path = os.path.join(folder_path, file_name)
        
        # Check if the file is an Excel file
        if file_name.endswith('.xlsx'):
            try:
                # Try to read the 'Top' and 'Bottom' sheets from the file
                excel_file = pd.ExcelFile(file_path)
                
                # Check if both 'Top' and 'Bottom' sheets exist
                if 'Top 10' in excel_file.sheet_names and 'Bottom 10' in excel_file.sheet_names:
                    top_data = excel_file.parse('Top 10')
                    bottom_data = excel_file.parse('Bottom 10')
                    
                    # Use full filename (without extension) as the strategy name
                    strategy_name = file_name.split('.')[0]  # Use the whole filename without the extension
                    top_data['Strategy'] = strategy_name
                    bottom_data['Strategy'] = strategy_name
                    
                    # Append the data from 'Top' and 'Bottom' sheets to the list
                    all_data.append(top_data)
                    all_data.append(bottom_data)
                else:
                    print(f"Skipping {file_name}: Missing 'Top' or 'Bottom' sheet.")
            except Exception as e:
                print(f"Error reading {file_name}: {e}")

    # Return a combined dataframe containing data from all valid files
    if all_data:
        return pd.concat(all_data, ignore_index=True)
    else:
        return pd.DataFrame()  # Return an empty DataFrame if no valid data was found


# Function to rank strategies by Sharpe Ratio and select the top and bottom 30%
def rank_strategies(data):
    # Sort by Sharpe Ratio (descending)
    sorted_data = data.sort_values(by='Sharpe Ratio', ascending=False)
    top_30_percent = sorted_data.head(int(len(sorted_data) * 0.3))  # Top 30% of data
    bottom_30_percent = sorted_data.tail(int(len(sorted_data) * 0.3))  # Bottom 30% of data

    return top_30_percent, bottom_30_percent


# Function to find consistent top and bottom performers across train and test datasets
def find_consistent_performers(train_data, test_data):
    # Rank both train and test data
    top_train, bottom_train = rank_strategies(train_data)
    top_test, bottom_test = rank_strategies(test_data)
    
    # Count occurrences of each strategy in the top 30% of both train and test
    top_train_counts = top_train['Strategy'].value_counts()
    top_test_counts = top_test['Strategy'].value_counts()
    
    # Count occurrences of each strategy in the bottom 30% of both train and test
    bottom_train_counts = bottom_train['Strategy'].value_counts()
    bottom_test_counts = bottom_test['Strategy'].value_counts()

    # Merge counts from train and test
    performance_data = pd.DataFrame({
        'Top Train': top_train_counts,
        'Top Test': top_test_counts,
        'Bottom Train': bottom_train_counts,
        'Bottom Test': bottom_test_counts
    }).fillna(0)

    # Add a column that calculates the "Total Occurrence" in top 30% and bottom 30%
    performance_data['Top Consistency'] = performance_data['Top Train'] + performance_data['Top Test']
    performance_data['Bottom Consistency'] = performance_data['Bottom Train'] + performance_data['Bottom Test']

    # Rank the strategies by "Top Consistency" (descending) and "Bottom Consistency" (ascending)
    top_performers = performance_data.sort_values(by='Top Consistency', ascending=False)
    bottom_performers = performance_data.sort_values(by='Bottom Consistency', ascending=False)

    return top_performers, bottom_performers


# Function to process all files, identify consistent performers, and save results
def process_results(train_folder, test_folder, output_file):
    # Read files from both train and test folders
    print("Reading train data...")
    train_data = read_files_from_folder(train_folder)
    print("Reading test data...")
    test_data = read_files_from_folder(test_folder)
    
    # Find consistent top and bottom performers
    print("Finding consistent performers...")
    top_performers, bottom_performers = find_consistent_performers(train_data, test_data)
    
    # Save results to an Excel file
    print(f"Saving results to {output_file}...")
    with pd.ExcelWriter(output_file) as writer:
        top_performers.to_excel(writer, sheet_name='Consistent Top Performers', index=True)
        bottom_performers.to_excel(writer, sheet_name='Consistent Bottom Performers', index=True)
    
    print("Process completed and results saved.")


# Main execution
if __name__ == "__main__":
    # Define the train and test folder paths
    train_folder = r"C:\Users\masho\OneDrive\Documents\Fall 24 SEM\Capstone\Final Results\BTC 2024 4hr TR"
    test_folder = r"C:\Users\masho\OneDrive\Documents\Fall 24 SEM\Capstone\Final Results\BTC 2024 4hr T"
    
    # Output file name
    output_file = 'consistent_performers.xlsx'
    
    # Run the process
    process_results(train_folder, test_folder, output_file)
```
